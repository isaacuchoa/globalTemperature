```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('globalTemperature.csv')
df.head(5)
```

```python
# Informações e tipo das colunas
df.info()
```

```python
# Mostrar os nomes das colunas
for col in df.columns:
    print(col)
```

```python
# Remover colunas com drop
df.drop(['Five-Year Anomaly',' Five-Year Unc.',' Ten-Year Anomaly',' Ten-Year Unc.','  Twenty-Year Anomaly',' Twenty-year Unc.'],axis = 1)


```

```python
df["Year"] = df["Year"] = 2023;

# Avaliar as anomalias por mês no ano 2023
sns.set_theme(style="ticks", palette="pastel")

plt.figure(figsize=(15,6))
plt.title("Análise de dados")


sns.boxplot(x="Year", y="Monthly Anomaly",
             palette=["m", "g"],
            data=df)
sns.despine(offset=10, trim=True);
```

```python
# Verificar outliers através de boxplot

plt.figure(figsize=(15,6))
plt.title('Analisando outliers')

sns.boxplot(data=df, x=' Monthly Unc.', y='Year');
```
